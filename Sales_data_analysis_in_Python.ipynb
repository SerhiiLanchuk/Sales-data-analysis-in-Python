```python
# Імпутація найбільш популярної країни для даного виду продукту.
events["country_code"] = events.groupby("product_id")["country_code"].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
```

```python
#Заповнення пропусків заглушкою "UNK" або "Unknown".
events["Country Code"] = events["Country Code"].fillna("Unknown")
```

```python
# Видалення пропущених значень.
events = events[~events["Country Code"].isnull()]
```

```python
# Заміна пропущених значень на медіану по категорії товару в колонці Units Sold.
events["units_sold"] = events.groupby("product_id")["units_sold"].transform(lambda x: x.fillna(x.median()))
```

Звіт містить опис процесу обробки даних, візуалізацію результатів, аналітичні висновки та бізнес-рекомендації.

Цей етап допоможе краще зрозуміти основні джерела прибутку, витрат і популярності товарів, а також визначити ефективність роботи на різних ринках і каналах продажу.

**Аналіз географічного розподілу продажів (по країнах):**

**Аналіз каналів продажу (онлайн/офлайн**):
- Дозволяє оцінити ефективність кожного каналу;

```python
#Переіменування колонки
countries.rename(columns={"name": "country_name"}, inplace=True)

#Об'єднання таблиць
df = events.merge(products, how="left", left_on="product_id", right_on="id")
df = df.merge(countries, how="left", left_on="country_code", right_on="alpha_3")
#Видалення колонок що повторюються
df.drop(columns=["id", "alpha_3"], inplace=True)
print(df)
```

```python
# Додаємо колонку прибутку
df["revenue"] = df["units_sold"] * df["unit_price"]
df["cost_total"] = df["units_sold"] * df["unit_cost"]
df["profit"] = df["revenue"] - df["cost_total"]

# Додаємо інтервал відвантаження
df["shipping_days"] = (df["ship_date"] - df["order_date"]).dt.days

print("Загальна кількість замовлень:", df["order_id"].nunique())
print("Загальний прибуток:", df["profit"].sum())
print("Загальний дохід:", df["revenue"].sum())
print("Кількість охоплених країн:", df["country_name"].nunique())
print("Кількість категорій товарів :", df["product_id"].nunique())
```

```python
#Аналіз за категоріями товарів
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Аналіз за категоріями товарів", fontsize=16)

# Доходи
sns.barplot(data=df, x="item_type", y="revenue", hue="item_type", estimator=sum, errorbar=None, legend=False,
            order=df.groupby("item_type")["revenue"].sum().sort_values(ascending=False).index, palette='viridis',
            ax=axes[0, 0])
axes[0, 0].set_title("Сумарні доходи")
axes[0, 0].set_xlabel("Категорія товару")
axes[0, 0].set_ylabel("Дохід")
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].grid(True, linestyle='--', alpha=0.5)

# Витрати
sns.barplot(data=df, x="item_type", y="unit_cost", hue="item_type", estimator=sum, errorbar=None, legend=False,
            order=df.groupby("item_type")["unit_cost"].sum().sort_values(ascending=False).index, palette='viridis',
            ax=axes[0, 1])
axes[0, 1].set_title("Сумарні витрати")
axes[0, 1].set_xlabel("Категорія товару")
axes[0, 1].set_ylabel("Витрати")
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(True, linestyle='--', alpha=0.5)

# Прибутки
sns.barplot(data=df, x="item_type", y="profit", hue="item_type", estimator=sum, errorbar=None, legend=False,
            order=df.groupby("item_type")["profit"].sum().sort_values(ascending=False).index, palette='viridis',
            ax=axes[1, 0])
axes[1, 0].set_title("Сумарний прибуток")
axes[1, 0].set_xlabel("Категорія товару")
axes[1, 0].set_ylabel("Прибуток")
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, linestyle='--', alpha=0.5)

# Популярність (за кількістю проданих одиниць)
sns.barplot(data=df, x="item_type", y="units_sold", hue="item_type", estimator=sum, errorbar=None, legend=False,
            order=df.groupby("item_type")["units_sold"].sum().sort_values(ascending=False).index, palette='viridis',
            ax=axes[1, 1])
axes[1, 1].set_title("Популярність товарів (шт.)")
axes[1, 1].set_xlabel("Категорія товару")
axes[1, 1].set_ylabel("Одиниці")
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

**Висновки**   

Розбіжність між обсягом і прибутковістю: Не завжди найпопулярніші товари (за кількістю проданих одиниць) приносять найбільший прибуток. Це вказує на те, що "*Beverages*" можуть мати низьку маржу прибутку.

Високомаржинальні товари: Категорія "*Cosmetics*" є ключовою для бізнесу, оскільки вона генерує найбільший прибуток, хоча її доходи та обсяги продажів (популярність) не є найвищими. Компанії слід зосередити увагу на цій категорії, щоб максимізувати прибутковість. Можливо, варто розглянути розширення асортименту або посилення маркетингу для "*Cosmetics*".

Офісні товари як основа бізнесу: Категорія "*Office Supplies*" є найбільш збалансованою і стабільною, оскільки вона лідирує як за доходами, так і за кількістю проданих одиниць, а також займає високе місце за прибутком. Це є "золотим активом" бізнесу, який забезпечує значну частину як загального доходу, так і прибутку.

Аналіз ефективності: Незважаючи на високі доходи та витрати, "*Household*" посідає третє місце за прибутком. Це може свідчити про потенціал для оптимізації. Компанії варто переглянути витрати, пов'язані з цією категорією, щоб підвищити її прибутковість.

Більшість товарів не мають чіткої сезонності. Продажі в багатьох категоріях, таких як "Baby Food", "Household", "Meat", "Office Supplies" та інші, є нерегулярними та не демонструють стабільних піків у певні місяці року.
